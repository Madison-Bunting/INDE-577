# Exploring a New Dataset

Before running analysis on a new dataset, one must first explore the data to gain an understanding of what they are working with. This notebook contains an example of doing this with the primary dataset used in this repository, [IBMEmployeeAttrition](https://raw.githubusercontent.com/Madison-Bunting/INDE-577/main/IBMEmployeeAttrition.csv). Originally, the dataset was hosted on IBM's website, but has since been taken down, however it can be found on Kaggle [here](https://www.kaggle.com/pavansubhasht/ibm-hr-analytics-attrition-dataset).

This fictional dataset contains integer variables that correlate with various qualitative responses, as though to a survey with response scales. The key for those numerical values is detailed below:
- Education: 1 = below college, 2 = college, 3 = bachelor, 4 = master, 5 = doctor
- Environment/Satisfaction: 1 = low, 2 = medium, 3 = high, 4 = very high
- Job Involvement: 1 = low, 2 = medium, 3 = high, 4 = very high
- Job Satisfaction: 1 = low, 2 = medium, 3 = high, 4 = very high
- Performance Rating: 1 = low, 2 = good, 3 = excellent, 4 = outstanding
- Relationship Satisfaction: 1 = low, 2 = medium, 3 = high, 4 = very high
- Work-Life Balance: 1 = bad, 2 = good, 3 = better, 4 = best


Those who are interested in investigating the intersection of HR and data science (often called "People Analytics") can find a list of other useful datasets [here](https://www.aihr.com/blog/hr-data-sets-people-analytics/).

```python
#Always include important imports at the top
import numpy as np #for linear algebra functions
import pandas as pd #for data processing the CSV

#visualization
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
#importing a dataset into a dataframe
df = pd.read_csv("https://raw.githubusercontent.com/Madison-Bunting/INDE-577/main/IBMEmployeeAttrition.csv"
                , index_col = 0)
df.head() #shows the first several rows of the dataset
```

```python
#size of dataset
df.shape
```

This dataset contains 1470 examples and 35 variables, including ID

```python
df.nunique().nsmallest(10)
```

```python
#Drop variables with zero variance: EmployeeCount, Over18, and StandardHours
df.drop('EmployeeCount', axis = 1, inplace = True)
df.drop('Over18', axis = 1, inplace = True)
df.drop('StandardHours', axis = 1, inplace = True)

#EmployeeNumber does not provide meaningful information for exploratory analysis, so we will drop it as well
df.drop('EmployeeNumber', axis=1, inplace=True)
#Replacing Yes/No in the target variable (Attrition) with 1/0
df['Attrition'] = df['Attrition'].map({'No':0, 'Yes':1})

#Confirm that those variables were dropped 
df.shape
```

```python
#Check for missing values
missing_df = df.isnull().sum()
print(missing_df)
```

No missing values were found

```python
#Check for duplicates
df[df.duplicated()]
```

No duplicates were found


## Visualization
After learning about the data, data scientists should start visualizing it to get a better sense of its shape and distributions. There are so many options of things to graph, but this dataset specifically examines employee attrition, so we start by simply graphing counts for those that have left vs those that have not.

```python
sns.set_style('darkgrid')
sns.countplot(x ='Attrition', data = df)
```

We see that the ratio of employees that leave to those that chose to stay is about 1:6

```python
# Attrition by Business Travel

df_attrition_y = df[df['Attrition']==1]
df_attrition_n = df[df['Attrition']==0]

fig, ax = plt.subplots(1,2)

fig.subplots_adjust(wspace=1)

sns.countplot(df_attrition_n['BusinessTravel'], ax=ax[0])
ax[0].set_title('Attrition = No')
plt.setp(ax[0].get_xticklabels(), rotation=90, ha="center")

sns.countplot(df_attrition_y['BusinessTravel'], ax=ax[1])
ax[1].set_title('Attrition = Yes')
plt.setp(ax[1].get_xticklabels(), rotation=90, ha="center")
```

These graphs demonstrate that employees who left were more likely to travel frequently than those who stayed.

*Note that the Travel_Frequently and Travel_Rarely values are switched between the graphs

```python
# Distribution of Monthly Income

tenure = sns.FacetGrid(data=df, col='Attrition', height=4, aspect=2)
tenure.map(plt.hist, 'MonthlyIncome', bins=150)

tenure.set(xlabel='Daily Rate', ylabel='No. of Employees')
plt.show()
```

Fewer employees who left had a high income.

The Kernel Density Estimate (KDE) plot below highlights this.

```python
# Monthly Rate distribution using Kernel Density Estimate (KDE) Plot

ax = sns.kdeplot(df.MonthlyIncome[(df["Attrition"] == 0) ],
                color="Green", shade = True)
ax = sns.kdeplot(df.MonthlyIncome[(df["Attrition"] == 1) ],
                ax =ax, color="Red", shade= True)
ax.legend(["Attrition = No","Attrition = Yes"], loc='upper right')
ax.set_ylabel('Density')
ax.set_xlabel('Monthly Income')
ax.set_title('Distribution of Monthly Income by Attrition')

ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
ax.spines['bottom'].set_linewidth(False)
ax.spines['left'].set_linewidth(False)

# figure size in inches
sns.set(rc={'figure.figsize':(7,5)})
```

This distribution of "Attrition = Yes" has a higher peak around lower monthly income, where as the distribution of "Attrition = No" is comparativley dense around higher income values.

```python
# Attrition by Department
fig, ax = plt.subplots(1,2)

fig.subplots_adjust(wspace=1)

sns.countplot(df_attrition_n['Department'], ax=ax[0])
ax[0].set_title('Attrition = No')
plt.setp(ax[0].get_xticklabels(), rotation=90, ha="center")

sns.countplot(df_attrition_y['Department'], ax=ax[1])
ax[1].set_title('Attrition = Yes')
plt.setp(ax[1].get_xticklabels(), rotation=90, ha="center")
```

These graphs demonstrate that employees who left were more likely to come from the Sales department than those who stayed.

*Note that the Travel_Frequently and Travel_Rarely values are switched between the graphs

```python
# Income levels by gender and marital status

fig, ax = plt.subplots(1,2)
fig.subplots_adjust(wspace=1)

sns.boxplot(x='Gender', y='MonthlyIncome', data=df, ax=ax[0])
sns.boxplot(x='MaritalStatus', y='MonthlyIncome', data=df, ax=ax[1])
```

Men and women have similar income levels, though single employees have lower income than those who are married and divorced. This is likely because the single employees are more likely to be in the early stages of their careers.

```python
#Visualize the distribution of the data by age and daily rate
sns.lmplot(x = 'Age', y = 'DailyRate', hue = 'Attrition', data = df) 
```

There does not appear to be a correlation in attrition value by daily rate or age.

```python
# Income levels by Education Field and Job Role
fig, ax = plt.subplots(2,1)
fig.subplots_adjust(hspace=0.33)

sns.boxplot(x='MonthlyIncome', y='EducationField', data=df, ax=ax[0])
sns.boxplot(x='MonthlyIncome', y='JobRole', data=df, ax=ax[1])
```

Employees seem to have similar income levels across education fields, though those from life sciences and medical backgrounds have a higher number of outliers.

Research Director and Manger are titles with the highest income levels, while Research Scientits, Laboratory Technician, and Sales Representative have the lowest.

This is dataset includes a large amount of information for each person. Because there could be other interesting relationships within the dataset, it is valuable to check for correlations.

```python
#Check for correlation
corr_matrix = df.corr()
corr_matrix
```

The next step is to visualize the correlations with a heatmap in Seaborn, but because this dataset contains a large number of columns, it will be divided into 3 heatmaps.

```python
#Dividing data into the corresponding heatmaps
map1 = corr_matrix.loc['Age':'NumCompaniesWorked', 'Age':'NumCompaniesWorked']
map2 = corr_matrix.loc['PercentSalaryHike':, 'Age':'NumCompaniesWorked']
map3 = corr_matrix.loc['PercentSalaryHike':, 'PercentSalaryHike':]

#Displaying the first map
sns.set(rc = {'figure.figsize':(15,8)})
sns.heatmap(map1, cbar=True, annot=True, square=True, fmt='.2f', annot_kws={'size': 11}, cmap="YlGnBu", 
            linewidths=0.5, linecolor='blue')
```

This map shows that Monthly Income and Job Level are highly correlated

```python
#Displaying the second map
sns.heatmap(map2, cbar=True, annot=True, square=True, fmt='.2f', annot_kws={'size': 11}, cmap="YlGnBu", 
            linewidths=0.5, linecolor='blue')
```

This map shows that Total Working Years is highly correlated with both Job Level and Monthly Income

```python
#Displaying the third map
sns.heatmap(map3, cbar=True, annot=True, square=True, fmt='.2f', annot_kws={'size': 11}, cmap="YlGnBu", 
            linewidths=0.5, linecolor='blue')
```

This map shows that Percent Salary Hike and Performance Rating are highly correlated, and that Years At Company is highly correlated with both Years in Current Role and Years with Current Manager. 

All of the insights from these heatmaps could be interesting things to investigate further and test with various machine learning techniques.
